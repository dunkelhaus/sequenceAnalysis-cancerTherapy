```python
#Mass vector
m = np.zeros((110))
m.fill(1.0/110.0)
m = m.reshape(110,1)
m
    
```

```python
#Centering Matrix
mt = m.T 
onem = np.ones((110,1))
rhs = onem * mt
lhs = np.eye(110)
centering = lhs - rhs
centering
```

```python
D = wdf.as_matrix()
ct = centering.T
centering = -0.5 *centering
product = centering.dot(D)
diff = product.dot(ct)
diff

#S = -0.5 * product
#S is the cross product matrix
```

```python
S = diff
S
```

```python
evals, evecs = np.linalg.eig(S)
diagonal = np.diag(evals)
diagonal.shape
```

```python
U = evecs
U
```

```python
diagonal
```

```python
mask = diagonal != 0
dist = diagonal[mask]
finalset = dist.reshape(1, dist.shape[0])
finalset
```

```python
total = finalset.sum()
total
```

```python
variances = ((1.0)/(float(total)))*finalset
variances
```

```python
percentages = variances * 100
percentages
```

```python
np.savetxt("fm_mutations_variances.txt", percentages, '%5.2f', '\n')
```
